```python
#importing required modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import cv2 as cv
import sklearn
```

```python
#accessing folders inside the specified path
folders=[]
DIR=r'dataset/train'
for folder in os.listdir(DIR):
    folders.append(folder)
folders
```

```python
classes=['No Glaucoma','Glaucoma']
```

```python
labels=[]
features=[]
```

```python
#creating x(features) and y(labels) data
for folder in folders:
    path=os.path.join(DIR,folder)
    for image in os.listdir(path):
        if folder=='class0':
            labels.append(0)
        else:
            labels.append(1)
        image_path=os.path.join(path,image)
        img=cv.imread(image_path)
        gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
        resized=cv.resize(gray,(290,290))
        features.append(resized)
```

```python
features
```

```python
labels
```

```python
#converting "features" into numpy array
x=np.array(features)
```

```python
x.shape
```

```python
#converting "labels" into numpy array
y=np.array(labels)
y.shape
```

```python
#creating x_test and y_test data
test_features=[]
test_labels=[]
DIR=r'dataset/test'
for folder in folders:
    path=os.path.join(DIR,folder)
    for image in os.listdir(path):
        if folder=='class0':
            test_labels.append(0)
        else:
            test_labels.append(1)
        image_path=os.path.join(path,image)
        img=cv.imread(image_path)
        gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
        resized=cv.resize(gray,(290,290))
        test_features.append(resized)
```

```python
test_features
```

```python
x_test=np.array(test_features)
x_test
```

```python
x_test.shape
```

```python
y_test=np.array(test_labels)
y_test
```

```python
y_test.shape
```

```python
#function to plot a sample data
def plot_sample(x,y,index):
    plt.imshow(x[index])
    plt.xlabel(classes[y[index]])
    plt.show()
```

```python
#reshaping images of 290*290 pixels into 84100
#scaling values by dividing with 255 in order to scale them between 0 and 1
scaled_x=x.reshape(-1,290*290)/255
scaled_x.shape
```

```python
scaled_x_test=x_test.reshape(-1,290*290)/255
scaled_x_test.shape
```

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score, recall_score, precision_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
```

```python
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(n_estimators=100, random_state=42)
ab = AdaBoostClassifier(n_estimators=100, random_state=42)
sv = SVC(probability=True)
```

```python

# Base learners
estimators = [
    ('lr', LogisticRegression(max_iter=1000)),
    ('rf_base', RandomForestClassifier(n_estimators=100)),
    ('dt', DecisionTreeClassifier(max_depth=5))
]

# Final estimator (meta-model)
final_estimator = RandomForestClassifier(
    n_estimators=100,        # Tune as needed
    max_depth=5,             # Control overfitting
    random_state=42
)

# Stacking classifier
hybrid_model = StackingClassifier(
    estimators=estimators,
    final_estimator=final_estimator,
    passthrough=False        # Set True to include original features if helpful
)

# # Fit the model
# hybrid_model.fit(scaled_x, y)

# # Evaluate
# y_pred_stack = hybrid_model.predict(scaled_x_test)
# accuracy_stack = accuracy_score(y_test, y_pred_stack)

# print("Accuracy:", accuracy_stack)
```

```python
hybrid_model.fit(scaled_x,y)
```

```python
#test data
ac=accuracy_score(hybrid_model.predict(scaled_x_test),y_test)
ac
```

```python
models = {
    'SVM': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'AdaBoost': AdaBoostClassifier(n_estimators=100),
    'Stacking Classifier': hybrid_model
}

# Initialize results list
results = []

# Evaluate each model
for name, model in models.items():
    model.fit(scaled_x,y)
    y_pred = model.predict(scaled_x_test)

    results.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
        'Recall': recall_score(y_test, y_pred, average='weighted'),
        'F1 Score': f1_score(y_test, y_pred, average='weighted')
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results,columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
print(results_df)
```

```python
def predict_and_plot(path):
    img=cv.imread(path)
    gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    resized=cv.resize(gray,(290,290))
    x_var=np.array(resized)
    x_var=x_var.reshape(-1,290*290)/255
    predicted=hybrid_model.predict(x_var)
    plt.imshow(img)
    plt.title(classes[predicted[0]])
    plt.show()
```
